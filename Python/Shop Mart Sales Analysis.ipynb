# Exploratory Data Analysis

### Is Unit Price affected by Customer Rating?

```python
sns.set_style("dark")
sns.color_palette()

g = sns.PairGrid(df_merged[['Unit price', 'Rating']], diag_sharey=False)
g.map_upper(sns.scatterplot, s=15)
g.map_lower(sns.kdeplot)
g.map_diag(sns.kdeplot, lw=2)

plt.show()
```

## Is there correlation between total price of purchase, tax 5%, cost of goods at point of sale, and gross income?

```python
sns.set(rc={"figure.figsize":(5, 5)})
ax = sns.heatmap(df_merged.corr(numeric_only=True).abs(),annot=False, linewidth=.5, cmap="crest")
plt.show()
```

### There is high correlation between these values as expected. We could only use one of these values when running the data.

## What is the average sales over time?

```python
%matplotlib inline
```

```python
df_merged['Date'] = pd.to_datetime(df_merged['Date'], errors='coerce')
df_merged["Time"] = pd.to_datetime(df_merged["Time"], errors='coerce')
```

```python
#grouping the data by month and then looking at the values in the total column 
df_merged.groupby(df_merged.Date.dt.month).mean(numeric_only=True)['Total Cost'].plot(kind='line', figsize=(22,5), marker='o')

xpoints =  [1,2,3,4,5,6,7]
ypoints =  np.around(df_merged.groupby(df_merged.Date.dt.month).mean(numeric_only=True)['Total Cost'].values,2)
for x, y in zip(xpoints, ypoints):
    plt.annotate(y,xy=(x, y), xytext=(-2, 8),textcoords='offset points', ha='center', va='bottom')

#adding chart title
plt.title('Month Wise Average Sales trend', fontsize=18)
#Removing top and right spines in the chart
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
#adding axis labels
plt.xlabel('Month')
plt.ylabel('Average Total Sales ($)')
#Setting the x axis ticks rotation to 0
plt.xticks(ticks =[1,2,3,4,5,6,7], labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'July'], rotation=0)
#Displaying the figure
plt.show()
```

### There is a downward trend in average total sales over the months

## What is the average COGS, Gross Margin, Gross Income over time?

```python
colors = plt.rcParams["axes.prop_cycle"]()

#making a list of the columns we want to plot
columns  = [ 'COGS', 'gross margin percentage', 'gross income']
for i in columns:  
    color = next(colors)["color"]
    #grouping the data by month and then plotting the columns above to find the trend over the time
    df_merged.groupby(df_merged.Date.dt.month).mean(numeric_only=True)[i].plot(kind='line', figsize=(22,5), color= color, marker='o')
    
    xpoints =  [1,2,3,4,5,6,7]
    ypoints =  np.around(df_merged.groupby(df_merged.Date.dt.month).mean(numeric_only=True)[i].values,2)
    for x, y in zip(xpoints, ypoints):
        plt.annotate(y,xy=(x, y), xytext=(-2, 8),textcoords='offset points', ha='center', va='bottom')
    
    #Adding Chart title
    plt.title(f'Month Wise Average {i}', fontsize=20)
    #Removing the top and right spines
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    #Adding axis labels
    plt.xlabel('Month')
    plt.ylabel(i)
    plt.xticks(ticks =[1,2,3,4,5,6,7],  labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'July'])
    #Displaying the chart
    plt.show()
```

### Gross income and COGS follow same trend as Average Sales Trend

## What are the different types of payment methods and breakdown % of them?

```python
labels = df_merged["Pay Method"].value_counts().index
sizes = df_merged["Pay Method"].value_counts()
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99',"pink","yellow"]
plt.figure(figsize = (16,16))
plt.pie(sizes, labels=labels, rotatelabels=False, autopct='%1.1f%%',colors=colors,shadow=True, startangle=45)
plt.title('Payment Method',color = 'red',fontsize = 15)
plt.show()  
```
